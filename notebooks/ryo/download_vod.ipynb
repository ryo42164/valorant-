```python
from pathlib import Path
from yt_dlp import YoutubeDL

def download_youtube_video(url, output_dir="data/vods"):
    output_path = Path(output_dir)
    output_path.mkdir(parents=True, exist_ok=True)

    ydl_opts = {
        "outtmpl": str(output_path / "%(title)s.%(ext)s"),
        "format": "mp4/bestvideo+bestaudio/best",
        "merge_output_format": "mp4",
    }

    with YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

url = "https://www.youtube.com/watch?v=72YPJbZQMdU"
#download_youtube_video(url)
```

```python

def list_formats(url):
    with YoutubeDL({}) as ydl:
        info = ydl.extract_info(url, download=False)
        for f in info.get("formats", []):
            print(
                f.get("format_id"),
                f.get("ext"),
                f.get("resolution"),
                f.get("vcodec"),
                f.get("acodec")
            )
```

```python
def download_youtube_video(url, output_dir="../data/vods"):
    output_path = Path(output_dir)
    output_path.mkdir(parents=True, exist_ok=True)

    ydl_opts = {
        "format": "299/398/303/bestvideo",
        "outtmpl": str(output_path / "%(title)s.%(ext)s"),
    }

    with YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

url = "https://www.youtube.com/watch?v=72YPJbZQMdU"
download_youtube_video(url,output_dir="../../data/vods")
```

```python
import pandas as pd
from pathlib import Path
df_path = Path("../data/raw/masters_santiago_2026_vods.csv")
vod_df = pd.read_csv(df_path)
vod_df[1:]["youtube_url"]
```

```python
saved_paths =[]
for url in vod_df[1:]["youtube_url"]:
    try:
        path = download_youtube_video(url, output_dir="../data/vods")
        saved_paths.append(path)
        print("downloaded:", path)
    except Exception as e:
        print("failed:", url, e)
```
