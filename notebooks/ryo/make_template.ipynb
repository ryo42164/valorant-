```python
from pathlib import Path
import cv2
import matplotlib.pyplot as plt

from valorant.video import get_frame_at_sec
from valorant.regions import crop_ui_region

vod_title = "M8 vs. EDG - VALORANT Masters Santiago - SWISS"
video_path = Path(f"../../data/vods/{vod_title}.mp4")

frame = get_frame_at_sec(video_path, 3000)
ui = crop_ui_region(frame)

plt.imshow(ui)
plt.axis("off")
plt.show()
```

```python
out_path = Path("../../data/templates/ui_round_Santiago.png")
out_path.parent.mkdir(parents=True, exist_ok=True)

cv2.imwrite(str(out_path), ui)
```
